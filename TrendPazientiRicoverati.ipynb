## Trend pazienti ricoverati/terapia intensiva
### Il codice genera un grafico che mostra l'andamento dei ricoverati e dei pazienti in terapia intensiva. E' stato scelto questo tipo di grafico in quanto i dati delle colonne HospitalizedPatients e IntensiveCarePatients  sono relative al bollettino giornaliero.

```python
    import pandas as pd
    import matplotlib.pyplot as plt

# Carica il dataset
    file_path_regione = 'covid19_italy_region _python.csv'
    data = pd.read_csv(file_path_regione)

# Converti la colonna "Date" in formato datetime e imposta come indice
    data['Date'] = pd.to_datetime(data['Date'])
    data.set_index('Date', inplace=True)

# Traccia il grafico
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['HospitalizedPatients'], label='Ricoverati', color='skyblue')
    plt.plot(data.index, data['IntensiveCarePatients'], label='Terapia intensiva', color='red')

    plt.title('Trend dei pazienti ricoverati e di quelli in terapia intensiva')
    plt.xlabel('Data')
    plt.ylabel('Numero di pazienti')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
# Salva il grafico come un'immagine PNG
    plt.savefig('TrendPazientiRicoverati.png')
    plt.show()
```
